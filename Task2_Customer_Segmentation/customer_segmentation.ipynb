Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/ifood_df.csv')
print(df.shape)
df.head()
```

Load and Explore Data

```python
df = pd.read_csv('/content/ifood_df.csv')
print(df.shape)
print(df.isnull().sum())
df.head()
df.describe()
```

Handle missing values

```python
df['Income'].fillna(df['Income'].median(),inplace=True)
```

Select features for clustering

```python
features = df[['Income','MntTotal','Age']]
print(features)
```

Scale the Data


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(features)
print(scaler)
```

Find best k using Elbow method

Looking at the graph 4 is K

```python
from sklearn.cluster import KMeans

inertia =[]
for k in range (1,11):
  km= KMeans(n_clusters=k,random_state=44)
  km.fit(scaled)
  inertia.append(km.inertia_)
plt.plot(range(1,11),inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Apply K-Means

```python
km =KMeans(n_clusters=4,random_state=42)
df['Clusters'] = km.fit_predict(scaled)
print(df['Clusters'].value_counts())
df.head()
```

Visualize cluasters

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Income'],y=df['MntTotal'],hue=df['Clusters'],palette='Set1')
plt.title('Customer Segments - income vs Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.show()
```

Analyze each cluster

```python
print(df.groupby('Clusters')[['Income','MntTotal','Age']].mean())
```
